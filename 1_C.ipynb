```python
import numpy as np
```

```python
M_ISS = 420e3          # kg
M_Z = 5.9722e24         # kg

r_Z = 6.3781e6          # m
h_ISS = 4.08e5         # m
kappa = 6.6743e-11      # m^3.kg^-1.s^-2

r_Z_ISS = r_Z + h_ISS

F_g = kappa * (M_ISS * M_Z) / r_Z_ISS**2
print(f"Gravitační síla mezi Země a ISS je: {F_g:.3e} N")
```

```python
v_ISS = np.sqrt(kappa * M_Z / r_Z_ISS)
r_Z_ISS = r_Z + h_ISS

F_d = (M_ISS * v_ISS**2)/r_Z_ISS
print(f"Vypočtená orbitální rychlost: {v_ISS:.2f} m/s")
print(f"Dostředivá síla působící na ISS je: {F_d:.3e} N")
```

```python
print(f"Poměr F_g / F_d: {F_g / F_d:.6f}")
```

```python
K = kappa * (M_Z/r_Z**2)
print(f"Intenzita gravitačního pole na povrchu země je: {K} N.kg^-1")
```

```python
K = kappa * (M_Z/r_Z_ISS**2)
print(f"Intenzita gravitačního pole na ISS je: {K} N.kg^-1")
```

```python
T = 23.93*60*60
omega = (2*np.pi)/T

F_g = kappa * (M_Z/r_Z**2)
F_o = 1*(omega**2)*r_Z

print(f"Intenzita tíhového pole na rovníku je: {F_g-F_o} N.kg^-1")
```

```python
r_Z = 6.3568*10**6

F_g = kappa * (M_Z/r_Z**2)
F_o = 1*(omega**2)*0
print(f"Intenzita tíhového pole na pólech je: {F_g-F_o} N.kg^-1")
```

```python
r_GEO = (kappa*M_Z/omega**2)**(1/3)
r_Z = 6.3781e6
h_GEO = r_GEO - r_Z

v = omega * r_GEO
print(f"Rychlost družice: {v:.0f} m/s")
print(f"Poloměr dráhy družice: {r_GEO/1e3:.0f} km")
print(f"Výška nad povrchem: {h_GEO/1e3:.0f} km")
```

```python
v1 = (kappa*M_Z/r_Z)**(1/2)
print(f"První kosmická rychlost: {v1:.0f} m/s")
v2 = (2*kappa*M_Z/r_Z)**(1/2)
print(f"Druhá kosmická rychlost: {v2:.0f} m/s")
```
