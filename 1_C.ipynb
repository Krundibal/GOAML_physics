```python
M_ISS = 420e3          # kg (420 tun)
M_Z = 5.972e24         # kg

r_Z = 6.378e6          # m
h_ISS = 4.08e5         # m
kappa = 6.674e-11      # gravitační konstanta

r = r_Z + h_ISS        # vzdálenost od středu Země

F_g = kappa * (M_ISS * M_Z) / r**2
print(f"Gravitační síla: {F_g:.3e} N")
```
